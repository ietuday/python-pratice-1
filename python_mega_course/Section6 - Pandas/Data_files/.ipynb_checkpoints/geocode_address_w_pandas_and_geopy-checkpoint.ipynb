```python
import pandas as pd
import geopy as geo
```

```python
df = pd.read_csv("supermarkets.csv")
df = df.set_index("ID")
df
```

```python
dir(geo)
```

```python
nom = geo.Nominatim()
nom.geocode("23 Brewster Ln E., Setauket, NY 11733")
```

```python
df["Address_Full"] = df.Address + ", " + df.City + ", " + df.State
df
```

```python
df["coordinates"] = df["Address_Full"].apply(nom.geocode)
df
```

```python
print(df.coordinates[1].latitude)
print(df.coordinates[1].longitude)
```

```python
df["Latitude"] = df["coordinates"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"] = df["coordinates"].apply(lambda x: x.longitude if x!= None else None)
```
